```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = salary_predict[['YearsExperience']]
y = salary_predict[['Salary']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf = LinearRegression()
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
plt.scatter(X_train,y_train,color='red',label='Training Data')
plt.plot(X_train,clf.predict(X_train),color='blue',label='Regression Line')
plt.title('Salary vs Experience(Training Data)')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
plt.scatter(X_test,y_test,color='green',label='Testing Data')
plt.plot(X_test,y_pred,color='blue',label='Regression Line')
plt.title('Salary vs Experience(Testing Data)')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
